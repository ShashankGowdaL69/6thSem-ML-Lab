```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/canada_per_capita_income.csv')

df = df.dropna()

X = df[['year']]
y = df['per capita income (US$)']

model = LinearRegression()
model.fit(X, y)

pred_2020 = model.predict([[2020]])

print("Predicted Per Capita Income in 2020:", pred_2020[0])

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Income vs Year")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/salary.csv')

df = df.dropna()

X = df[['YearsExperience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

pred_salary = model.predict([[12]])

print("Predicted Salary for 12 years experience:", pred_salary[0])

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/hiring.csv')

word_to_num = {
    "zero":0, "one":1, "two":2, "three":3, "four":4,
    "five":5, "six":6, "seven":7, "eight":8,
    "nine":9, "ten":10, "eleven":11, "twelve":12
}

df['experience'] = df['experience'].replace(word_to_num)

df['experience'] = pd.to_numeric(df['experience'], errors='coerce')
df['test_score(out of 10)'] = pd.to_numeric(df['test_score(out of 10)'], errors='coerce')
df['interview_score(out of 10)'] = pd.to_numeric(df['interview_score(out of 10)'], errors='coerce')
df['salary($)'] = pd.to_numeric(df['salary($)'], errors='coerce')

df['experience'] = df['experience'].fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())

X = df[['experience','test_score(out of 10)','interview_score(out of 10)']]
y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

p1 = model.predict([[2,9,6]])
p2 = model.predict([[12,10,10]])

print("Predicted Salary for (2 yr, 9 test, 6 interview):", p1[0])
print("Predicted Salary for (12 yr, 10 test, 10 interview):", p2[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/1000_Companies.csv')

df = pd.get_dummies(df, drop_first=True)

X = df.drop('Profit', axis=1)
y = df['Profit']

model = LinearRegression()
model.fit(X, y)

sample = pd.DataFrame([[91694.48, 515841.3, 11931.24, 1, 0]],
columns=X.columns)

pred_profit = model.predict(sample)

print("Predicted Profit:", pred_profit[0])
```
